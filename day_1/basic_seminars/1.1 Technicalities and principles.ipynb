```python
# http://scikit-learn.org/stable/auto_examples/cluster/plot_cluster_iris.html
%pylab inline --no-import-all
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

np.random.seed(5)

centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target
y = np.choose(y, [1, 2, 0]).astype(np.float)
```

```python
fig = plt.figure()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length\n\n')
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, s=200)
plt.subplots_adjust(top=0.5) 
```

# Essense - fitting a complex model on data with unknown dependence.

```python
np.random.seed(231)
min_x = 0
max_x = 1
x = np.random.uniform(min_x, max_x, size=20)
x_space = np.linspace(min_x, max_x, num=100)
y1_points = x + np.random.normal(0, 0.1, size=x.shape)
y2_points = np.sin(10*x) * np.exp(np.sqrt(x)) + (x+0.3)*np.random.normal(0, 1, size=x.shape)
y1_fit = np.polyfit(x, y1_points, deg=2)
y2_fit = np.polyfit(x, y2_points, deg=2)
```

```python
fig = pl.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 2, 1)
ax.scatter(x, y1_points)
ax.plot(x_space, np.poly1d(y1_fit)(x_space), 'r')
ax.set_title("Good model")
ax = fig.add_subplot(1, 2, 2)
ax.scatter(x, y2_points)
ax.plot(x_space, np.poly1d(y2_fit)(x_space), 'r')
_ = ax.set_title("Bad model")
```
