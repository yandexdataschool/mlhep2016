## Pandas
1. High level library with versatile data structures buit on top of NumPy
2. We'll only cover very basics - for more you can use https://bitbucket.org/hrojas/learn-pandas

```python
import root_numpy
events = root_numpy.root2array("public_train_10000.root", stop=10)
```

```python
import pandas as pd
import numpy as np
```

The main data structure is a data frame. It can be created from a ndarray of named fields

```python
events_pd=pd.DataFrame(events)
events_pd
```

Or you can manually pass the column names

```python
random_data = np.random.normal(size=(10, 5))
random_data.shape
```

```python
pd.DataFrame(random_data, columns=("i", "d", "d", "q", "d"))
```

columns can be nicely accessed

```python
events_pd.lepton_pt
```

and rows (more or less as in NumPy)

```python
events[0]
```

and in the same time

```python
events_pd[["lepton_pt", "mem_pt"]][0:5:2]
```

A convinient routine to have an overview. More on plotting later.

```python
%matplotlib inline
events_pd.hist(figsize=(15, 15));
```

Saving and loading files

```python
events_pd.to_csv("events.csv", index=False)
```
